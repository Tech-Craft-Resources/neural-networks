```python
import keras
from keras.applications import VGG16
```

```python
model = VGG16(weights='imagenet')
```

```python
model.summary()
```

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from keras.applications.vgg16 import decode_predictions, preprocess_input
from keras.utils import img_to_array, load_img

IMAGE_EXTENSIONS = {".jpg", ".jpeg", ".png", ".avif", ".webp", ".gif", ".bmp"}

# Carpeta images: notebook en ejercicios-nn/ → ../images; si el cwd es la raíz del repo → images/
for candidate_dir in (Path("..") / "images", Path("images")):
    if candidate_dir.is_dir():
        IMAGES_DIR = candidate_dir.resolve()
        break
else:
    raise FileNotFoundError(
        "No se encontró la carpeta images (prueba ../images desde ejercicios-nn/ o images/ en la raíz)"
    )

image_paths = sorted(
    p
    for p in IMAGES_DIR.iterdir()
    if p.is_file() and p.suffix.lower() in IMAGE_EXTENSIONS
)

if not image_paths:
    raise FileNotFoundError(f"No hay imágenes reconocidas en {IMAGES_DIR}")

for IMG_PATH in image_paths:
    img_pil = load_img(IMG_PATH, target_size=(224, 224))
    plt.figure(figsize=(6, 6))
    plt.imshow(img_pil)
    plt.axis("off")
    plt.title(f"{IMG_PATH.name} → 224×224 RGB")
    plt.show()

    x = img_to_array(img_pil)
    x = np.expand_dims(x, axis=0)
    # Preprocesado ImageNet para VGG16 (centrado por canal BGR / RGB según convención Keras)
    x_preprocessed = preprocess_input(x.copy())

    preds = model.predict(x_preprocessed, verbose=0)
    print(f"Top-5 ImageNet ({IMG_PATH.name}):")
    for _, label, score in decode_predictions(preds, top=5)[0]:
        print(f"  {label}: {score:.4f}")
    print()
```
